```python
# getting count of age values
data.age.count()
```

```python
#using interpolate method to replace null values of numeric column
data['age']=data['age'].interpolate(inplace=False)
```

```python
data
```

```python
# replacing NaN values with NA using replace method
data['gender']=data['gender'].replace(np.nan,'NA',inplace=False)
```

```python
data
```

```python
# Converting object in string 
data['attendance']=data['attendance'].astype('string')
data.info()
```

```python
# Converting string into int
data['attendance']=data['attendance'].str.replace("%","")
data
```

```python
data['attendance']=data['attendance'].astype('int64')
```
